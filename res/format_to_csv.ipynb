```python
import pandas as pd
```

```python
df = pd.read_table("optimaDataset_14.dat")
```

Rearranging columns inorder to have 'Choice' field in column 1

```python
df_columns = df.columns.tolist()
df_columns.remove('Choice')
df_columns.insert(1, 'Choice')
df_columns.remove('ReportedDuration')
df_columns.insert(2, 'ReportedDuration')
df_columns.remove('InVehicleTime')
df_columns.insert(3, 'InVehicleTime')
df = df[df_columns]
```

```python
df.to_csv("optimaDataset_14.csv")
print(df.shape)
df.head()
```

Removing rows whose choice value is NaN (-1)

```python
df = df[df['Choice'] != -1]
print(df.shape)
df.head()
```

```python
df.to_csv("optimaDataset_14_NaNChoice_dropped.csv")
```
